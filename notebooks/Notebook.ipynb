```python
import pandas as pd

df = pd.read_csv('C:\\Users\\Admin\\Desktop\\BCG GEN AI\\Boston_consulting_group_Job_simulation\\data\\Financial data - Sheet1.csv')

# Check data
print(df.head())
print(df.dtypes)
```

```python
cols = ['Total Revenue', 'Net Income', 'Assets', 'Liabilties', 'Operating Cash flow']

for col in cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df = df.sort_values(by=['Company', 'Year'])

df['Revenue Growth (%)'] = df.groupby('Company')['Total Revenue'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby('Company')['Net Income'].pct_change() * 100
```

```python
# Sort the dataframe by Company and Year
df = df.sort_values(by=['Company', 'Year'])

# Calculate Revenue Growth %
df['Revenue Growth (%)'] = (
    df.groupby('Company')['Total Revenue']
    .pct_change(fill_method=None) * 100
)

# Calculate Net Income Growth %
df['Net Income Growth (%)'] = (
    df.groupby('Company')['Net Income']
    .pct_change(fill_method=None) * 100
)

# Display results
print(df[['Company', 'Year', 'Total Revenue', 'Revenue Growth (%)',
          'Net Income', 'Net Income Growth (%)']])
```
